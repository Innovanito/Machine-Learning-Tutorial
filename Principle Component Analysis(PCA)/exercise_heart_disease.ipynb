```python
import pandas as pd
import numpy as np

df=pd.read_csv('heart.csv')
df.head(20)
```

```python
df.std()
```

```python
df.describe()
```

```python
df.keys()
```

```python
features = ['RestingBP', 'Cholesterol', 'FastingBS', 'MaxHR', 'Oldpeak']
```

```python
df[df.Cholesterol > (df.Cholesterol.mean() + 3*df.Cholesterol.std()) ]
```

```python
df1 = df

for feature in features:
  df1 = df1[df1[feature] <= (df1[feature].mean() + 3*df1[feature].std() )]
  print(df1.shape)
  df1 = df1[df1[feature] >= (df1[feature].mean() - 3*df1[feature].std())]
  print(df1.shape)

df_new =df1.copy()
```

```python
df1.head(30)
```

```python
df1.ChestPainType.unique()
```

```python
df1.RestingECG.unique()
```

```python
df1.ST_Slope.unique()
```

```python
df2 = df1.copy()
df2.ExerciseAngina.replace(
  {
    'N': 0,
    'Y': 1
  },
  inplace=True
)

df2.ST_Slope.replace(
  {
    'Down' : 1,
    'Flat' : 2,
    'Up' : 3
  },
  inplace=True
)

df2.RestingECG.replace(
  {
    'Normal': 1,
    'ST': 2,
    'LVH': 3
  },
  inplace=True
)

df2.head()
```

```python
df3 = pd.get_dummies(df2, drop_first=True)
df3.head()
```

```python
inputs = df3.drop('HeartDisease', axis=1)
outputs = df3['HeartDisease']
inputs
```

```python
outputs
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
inputs_scaled = scaler.fit_transform(inputs)
inputs_scaled
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(inputs_scaled, outputs, test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
model.score(X_test, y_test)
```

```python
from sklearn import svm

model2 = svm.SVC()
model2.fit(X_train, y_train)
model2.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier

model3 = RandomForestClassifier()
model3.fit(X_train, y_train)
model3.score(X_test, y_test)
```
