```python
import pandas as pd
df = pd.read_csv('titanic.csv')
df.head()
```

```python
inputs = df.drop('Survived', axis='columns')
output = df['Survived']
```

Delete the column value of PassengerId,Name,Ticket,Cabin,Embarked


Then us LabelEncoder and fit_transform to chagne Sex and Embarked values into number

```python
inputs_2 = inputs.drop(['PassengerId','Name', 'Ticket', 'Cabin', 'Fare'], axis='columns')
inputs_2
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_sex = LabelEncoder()
le_embarked = LabelEncoder()
```

```python
inputs_2['Sex_n']=le_sex.fit_transform(inputs_2['Sex'])
inputs_2['Embarked_n']=le_embarked.fit_transform(inputs_2['Embarked'])
```

```python
inputs_2[:20]
```

```python
inputs_2 = inputs_2.drop(['Sex', 'Embarked'], axis='columns')
inputs_2[:20]
```

```python
inputs_2
```

Now I have to deal with how to change NAN value into some value

I think I am going to mean all the age value and replace NAN value into mean value

```python
inputs_2 = inputs_2.fillna(0)
inputs_2[:20]
```

```python
inputs_2.mean()
```

```python
inputs_2.median()
```

```python
inputs_2['Age'] = inputs_2['Age'].replace(0,inputs_2['Age'].mean())
inputs_2[:20]
```

```python
inputs_2['Age'][:20].interpolate()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs_2, output, test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
