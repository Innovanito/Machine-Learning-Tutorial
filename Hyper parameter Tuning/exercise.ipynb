```python
from sklearn.datasets import load_digits
digits = load_digits()
```

```python
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.tree import DecisionTreeClassifier
```

```python
inputs = digits.data
outputs = digits.target
outputs
```

```python
import pandas as pd

coloumn_names=digits.feature_names
df = pd.DataFrame(inputs, columns=coloumn_names)
df
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
model1 = svm.SVC()
model2 = RandomForestClassifier()
model3 = LogisticRegression()
model4 = GaussianNB()
model5 = MultinomialNB()
model6 = DecisionTreeClassifier()
models = [model1, model2, model3, model4, model5, model6]
```

```python
model_params = {
  'svm': {
    'model': svm.SVC(gamma='auto'),
    'params': {
      'C': [1, 10, 20],
      'kernel': ['rbf', 'linear']
    }
  },
  'random_forest': {
    'model': RandomForestClassifier(),
    'params': {
      'n_estimators': [10, 50 ,100]
    }
  },
  'logistic_regression': {
    'model': LogisticRegression(solver='liblinear', multi_class='auto'),
    'params': {
      'C': [1, 5, 10]
    }
  },
  'GaussianNB': {
    'model': GaussianNB(),
    'params': {
      'var_smoothing': [1e-09, 1e-08, 1e-07]
    }
  },
  'MultinomialNB': {
    'model': MultinomialNB(),
    'params': {
      'alpha': [1, 2, 3]
    }
  },
  'DecisionTreeClassifier': {
    'model': DecisionTreeClassifier(),
    'params': {
      'criterion': ['gini', 'entropy']
    }
  }
}
```

```python
scores = []
cv_result ={}

for model_name, mp in model_params.items():
  clf = GridSearchCV(mp['model'], mp['params'], cv=8, return_train_score=True)
  clf.fit(inputs, outputs)
  scores.append({
    'model': model_name,
    'best_score': clf.best_score_,
    'best_params': clf.best_params_,
  })
  cv_result[model_name] = clf.cv_results_
```

```python
df_result = pd.DataFrame(scores, columns=['model', 'best_score', 'best_params'])
df_result
```

```python
for model in cv_result:
  print(dir(cv_result[model]))
```
